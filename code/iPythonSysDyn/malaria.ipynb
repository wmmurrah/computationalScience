### Program to run simulation of the spread of malaria

```python
#This code performs the necessary steps to prepare to plot.
from pylab import *     #Import plotting module matplotlib as well as other modules
                        #into the global namespace.

#Set the default plot to be inline rather than a new window.
%matplotlib inline

import inspect  #This allows us to print the source code in this notebook.
from malaria import *  #Import the custom functions from this project.

goldenratio=1/2*(1+sqrt(5))        #The next few lines are used for the size of plots
fsx=7                              #Width (in inches) for the figures.
fsy=fsx/goldenratio                #Height (in inches) for the figures.
```

```python
#Simulate
tLst, uninfected_humansLst, vectorsLst, uninfected_mosquitoesLst, human_hostsLst, immuneLst = malaria()
```

```python
#Graph
figure(figsize=(fsx, fsy))
plot(tLst, uninfected_humansLst, 'b-.', label='Uninfected Humans')
plot(tLst, vectorsLst, 'k-', label='Vectors')
plot(tLst, uninfected_mosquitoesLst, 'k-.', label='Uninfected Mosquitoes')
plot(tLst, human_hostsLst, 'b-', label='Hosts')
plot(tLst, immuneLst, 'g-', label='Immune')
legend(loc='center right')
xlabel('Time')
ylabel('Population')
title('Populations vs Time')
show()
```
